```python
import sys
sys.path.append('/media/max/TOSHIBA EXT/ONERA_SONDRA/algos/contrastive_learning')
from pipeline.storage.state import StateStorageFile, StateStorageBase
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def plot_param(dic:StateStorageBase,key:str or list,epoch=None,step_per_epoch=10):
    n_step = dic.get_value('n_step')
    if type(key) == str:
        key = [key]
    for k in key:
        if not dic.has_key(k):
            print('Parameter {} not in dic'.format(k))
            return False
        l = dic.get_value(k)
        step = np.array(list(range(len(l))))/n_step
        step_sz = int(n_step/step_per_epoch)
        if epoch == None:
                l = l[::step_sz]
                step = step[::step_sz]
        else :
            l = l[epoch[0]:epoch[1]:step_sz]
            step = step[epoch[0]:epoch[1]:step_sz]
        plt.plot(step,l,label=k)

    plt.legend()
    plt.show()
```

```python
dic = StateStorageFile('/media/max/TOSHIBA EXT/ONERA_SONDRA/algos/contrastive_learning/pipeline/out/contrastive_onera_16/tracker')
```

```python
for key in dic._state:
    print(key)
```

```python
plot_param(dic,'lr',step_per_epoch=1)
```
